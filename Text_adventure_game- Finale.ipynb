```python

# Writing the doctype text
"""
It’s 5pm and you just finished from school and your dad comes by to pick you up from school with your mom. 
You enter the car and get going and suddenly while singing to one of your best songs from the radio, 
a car attacks you guys from behind and your dad loses control of the wheels and drives 11km into the woods. 
Boom!!! There was a car crash, and both your mom and dad didn’t make it.

You managed to escape the crash before the explosion and now you Must find your way back to the road before
nightfall and get to the nearest hospital for check up and to get yourself treated. 
There’s no network there, so you can’t use your cell phone.
          
         
"""
# Importing the time function. 
import time


delay = 0.3 # delay time set to 0.3


#Introduction
name = input(prompt = "what is your name?\n")

# printing dynamically to allow for inputs in our text
print(f"""
-------------------------------------------------------------------------------------------------------------

Welcome {name}! 
It’s 5pm and you just finished from school and your dad comes by to pick you up from school with you mom. 
You enter the car and get going and suddenly while singing to one of your best songs from the radio, 
a car attacks you guys from behind and your dad loses control of the wheels and drives 11km into the woods. 
Boom!!! There was a car crash, and both your mom and dad didn’t make it. 

You managed to escape the crash before the explosion and now you MUST find your way back to the road before 
nightfall and get to the nearest hospital for check up and to get yourself treated. 
There’s no network there, so you can’t use your cell phone. Let's get you to the hospital.

-------------------------------------------------------------------------------------------------------------
""")
#This will slow down the time it takes for the input function to execute
time.sleep(delay)

#start of the game
def start_game():
    answer = input(prompt= "Type Y or N to start game? \n").casefold()
    
    # conditional statement for if the answer is 'no'
    if 'n' in answer or 'N' in answer or 'no' in answer:
        print('Try again next time!')
        exit()
    
    # conditional statement for if the answer is 'yes'
    elif 'es' in answer or 'y' in answer:
        print('Good job! your journey begins now!') 
        level_1()
        


# defining the first level
def level_1():
    
    # Let's print some texts to follow through with the story
    print(f"""
------------------------------------------------------------------------------------------------------------    
    You leave the crash scene and start walking towards the road. After walking 
    some distance, you meet a 'Y' junction.
    
    You have two options:
    1. Go right 
    2. Go left
------------------------------------------------------------------------------------------------------------        
    """)
    # Giving the user an input box
    choice = input(prompt= "which way would you go, right or left?\n").casefold() # the .casefold allows for both upper and lower case from the user
    
    # writing a block of 'if' and 'elif'
    if choice == '1':
        print("Alright! It's about time!")
        level_2()
        
    elif 'go right' in choice or 'right' in choice or 'r' in choice:
        print("Alright! It's about time!")
        level_2()
        
    elif choice == '2':
        print("Try again!")
        input(prompt = "press enter to continue. \n")
        level_1()
        
    elif 'go left' in choice or 'left' in choice or 'l' in choice:
        print("Try again!")
        input(prompt = "press enter to continue. \n")
        level_1()
        
    
    # A conditional statement for if the user mistakenly or intentionally presses the enter key without a value     
    elif choice == (""):
        print("You must select one option!")
        level_1()
    
    # Just because having an else clause after an if statement is good practice
    else:
        print('if you chose to play the game, play by the rules! Now choose an option. \n')
        
# defining the fail function
def fail():
    print("You were killed by the bear!")
    level_2()
    
# Defining the play again function    
def play_again():
    
    choice = input(prompt = "Do you want to play again? Y/N \n").casefold()
    
    if 'y' in choice or 'Y' in choice:
        level_1()
        
    elif 'n' in choice or 'N' in choice:
        exit()
        
    else:
        print("Are you sure about your decision? Try the letters Y/N!")
        play_again()
    
    
# if you decided to go right from level one, it will run the following codes 
#till it gets to "if you decided to go left from level_1" 



    
#if you decided to go left from level_1 
time.sleep(delay)

# Defining level_2
def level_2():
    # Printing some part of the story for follow up
    print(f"""
----------------------------------------------------------------------------------------------------------    
    After walking some distance, you see a medium-sized toolbox and in it contains some items you will need
    for your journey to the hospital: a first aid box, a 9mm pistol, a pocket knife, a small pack of carrots, 
    torchlight and a long jacket.
    
    You have two options:
    1. Pick only the items needed
    2. Carry the entire box.
----------------------------------------------------------------------------------------------------------    
    """)
    # Giving the user an input box
    choice = input(prompt = "which option will you choose?\n").casefold()
    
    # Adding a try except clause
    
        
    try:
            
        choice = int(choice)
            
        # Writing a block of if and elif statement
        if choice == 1:
                           
            print("Great decision! Let's get to the road")
            level_3()

        elif choice == 2:

            while choice == 2:

                print("that's not the correct decision!. The box will weigh you down. You need to hurry!")
                    
                input(prompt = "Try again! press enter to continue.\n")
                level_2()
                    
               
    except ValueError:
        print("oops! That's an invalid number. Try again!")
        input(prompt = "Try again! press enter to continue.\n")
        level_2()
            
time.sleep(delay)          
            
def level_3():
    # Printing some part of the story for follow up
    print(f"""
-----------------------------------------------------------------------------------------------------------    
    After walking for a while, you've found a spring to drink water from to keep you hydrated for your journey.
    Do you want to:
    1. Keep walking
    2. Drink from the spring
-----------------------------------------------------------------------------------------------------------    
    """)
    # Giving the user an input box
    choice = input(prompt = "what do you want to do?\n").casefold()
    
            
    try:
            
        choice = int(choice)

        if choice == 1:
            print("You died! You've been walking for so long dehydrated! \n")
            fail()
            
        elif choice == 2:
            print("Well done. You've gained more energy for the rest of the journey!")
            level_4()
            
            
    except ValueError:
        print("oops! That's an invalid number. Try again!")
        input(prompt = "press enter to continue! \n")
        level_3()
            
            
time.sleep(delay)

#Defining level_4
def level_4():
    # Printing some part of the story for follow up
    print(f"""
    ----------------------------------------------------------------------------------------------------
    You've walked some distance and now you’re hungry and want something to eat. 
    You find a rabbit running around.
    What would you do?
    1. Lure the Rabbit with some carrots
    2. Shoot the rabbit
    ----------------------------------------------------------------------------------------------------
    """)
    # Giving the user an input box
    choice = input(prompt = "What will you do? \n").casefold()
    
    # Adding a try except clause
            
    try:
      
        choice = int(choice)
        if choice == 1:
            print("Good! You've killed the rabbit. Now you can have dinner and proceed!")
            level_5()
           
        elif choice == 2:   
            print("Bad decision! You scared the rabbit off!")
            input(prompt = "press enter to continue! \n")
            level_4()
                
        
    except (TypeError, ValueError):
        print("oops! That's an invalid number. Try again!")
        input(prompt = "press enter to continue! \n")
        level_4()


# Defining the final level of the game.
time.sleep(delay)
def level_5():
    # Printing some part of the story for follow up
    print(f"""
--------------------------------------------------------------------------------------------------------    
    As you walk in the woods, you hear the rustling of the trees and get scared. You start to run
    until you get to a bridge. The bridge is your only way to the road but you've got a bear right 
    in front of the bridge.
    
--------------------------------------------------------------------------------------------------------    
    """)
    
    # Making the player guess the correct number. They have three tries to get the correct number from 1-5
    
    # The three options to choose from are:
    # 1. Run away
    # 2. Do nothing
    # 3. Shoot the Bear
    # The correct answer is the third option
    
    import random # to execute the guessing of a number, we import a random function from python's library
    
    number = 3 # the correct random number 
    tries = 0  # the number of tries
    
    # Adding a try except clause and a while true loop
    while True:
        choice = input(prompt = "What do you do? Guess a number from 1-10 \n").casefold()
        tries += 1
        
        try:
            choice = int(choice)
        
            if choice == number:
                print(f"""Congratulations {name}! You shot the bear and now you're on your way to the hospital. \n""")
                play_again()
                break
                
                       
            
            elif choice != number:
            
                if choice > number:
                    print("Your guess is too high! Try again. \n")
                
                elif choice < number:
                    print("Try again!")
        except ValueError:
            print("Oops! I need a number! \n")
            input("Press enter to continue! \n")
            
  
    
start_game()   

        
        

```
